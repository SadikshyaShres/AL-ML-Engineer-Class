Slicing String :You can return a range of characters by using the slice syntax.Specify the start index and the end index, separated by a colon, to return a part of the string.

string[start:stop:step]

start: where slicing begins,
stop: where slicing ends, but this position is not included,
step: how many positions to move


```python
b = "Hello, World!"
print(b[2:5])

#slicing from start

b = "Hello, World!"
print(b[:5])

#slicing from end 

b = "Hello, World!"
print(b[2:])
```

```python
#Negative indexing 
b = "Hello, World!"
print(b[-5:-2])
```

LIST: 

List is collection used to store multiple values in one variable.

Lists are written using square brackets [], and items are separated by commas.

Properties of lists

Python lists are:

Ordered: Every item has a fixed index position.

Mutable: Items can be added, removed, or changed.

Indexed: The first item's index is 0.

Allow duplicates: The same value can appear multiple times.

Allow different data types: One list can contain strings, numbers, and Boolean values.

```python
name = "Python"                  # String
languages = ["Python", "SQL"]  # List
```

```python
#indexing 

fruits = ["apple", "banana", "mango", "orange"]

print(fruits[0])
print(fruits[1])
print(fruits[3])
```

```python
#Negative indexing 

fruits = ["apple", "banana", "mango", "orange"]

print(fruits[-1])
print(fruits[-2])
```

```python
#slicing 

fruits = ["apple", "banana", "mango", "orange", "grapes"]

print(fruits[1:4])
```

```python
#slice to end 
print(fruits[2:])

```

```python
#reverse a list with slicing

print(fruits[::-1])
```

```python
#list is mutable meaning their content can be changed after creation 

fruits = ["apple", "banana", "mango"]

fruits[1] = "orange"

print(fruits)
```

```python
#.Append()- add one item to the end of the list 

fruits = ["apple", "banana"]

fruits.append("mango")

print(fruits)
```

```python
numbers = [1, 2]

numbers.append([3, 4])

print(numbers)
```

```python
#.pop() - removes and returns a item 

fruits = ["apple", "banana", "mango"]

removed_item = fruits.pop()

print(fruits)
print("Removed:", removed_item)
```

```python
fruits = ["apple", "banana", "mango"]

removed_item = fruits.pop(1)

print(fruits)
print("Removed:", removed_item)
```

```python
#extend()- adds every item from another collection to the end of the list.

fruits = ["apple", "banana"]
more_fruits = ["mango", "orange"]

fruits.extend(more_fruits)

print(fruits)
```

```python
#using append - adds the complete value as one item.
numbers = [1, 2]
numbers.append([3, 4])

print(numbers)

#using extend- adds each value separately.

numbers = [1, 2]
numbers.extend([3, 4])

print(numbers)

```

```python
#remove()- remove the first matching value 

fruits = ["apple", "banana", "mango", "banana"]

fruits.remove("banana")

print(fruits)
```

```python
#reverse - reverses the original list.

numbers = [1, 2, 3, 4]

numbers.reverse()

print(numbers)
```

```python
#sort()- arranges the original list in ascending order.

numbers = [40, 10, 30, 20]

numbers.sort()

print(numbers)
```

```python
#sort in descending order
numbers = [40, 10, 30, 20]

numbers.sort(reverse=True)

print(numbers)
```

```python
#Nested list- is a list inside another list

students = [
    ["Sadikshya", 90],
    ["Joey", 85],
    ["Monica", 95]
]

print(students[0])

print(students[0][0])
print(students[0][1])
```

```python
#changing the value in a nested list
students[1][1] = 88

print(students)
```

```python
#Loop through a nested list 
for student in students:
    print("Name:", student[0])
    print("Score:", student[1])
```

List Comprehension - is a short way to create a new list using a loop 



```python
squares = []

for number in range(1, 6):
    squares.append(number ** 2)

print(squares)
```

```python
#using list comprehension 

squares = [number ** 2 for number in range(1, 6)]

print(squares)
```

```python
#list comprehension using if-else 

numbers = [1, 2, 3, 4, 5]

results = ["Even" if number % 2 == 0 else "Odd" for number in numbers]

print(results)
```

Tuple: 

Tuple is a collection used to store multiple values in one variable.

Tuples are usually written using parentheses (), and their items are separated by commas.

Properties of a Tuple

Ordered: Every item has a fixed position.

Indexed: Indexing begins at 0.

Immutable: Items cannot be added, removed, or changed after creation.

Allow duplicates: The same value can appear multiple times.

Allow different data types: A tuple can contain strings, numbers, Boolean values, lists, and other tuples.

```python
colors = ("red", "green", "blue")
```

```python
#tuple without parenthesis 

colors = "red", "green", "blue"

print(colors)
print(type(colors))
```

```python
fruit = ("apple",) #tuple with one item 

print(type(fruit))
```

```python
#tuple indexing 

fruits = ("apple", "banana", "mango", "orange")

print(fruits[0])
print(fruits[2])
```

```python
#negative indexing 

fruits = ("apple", "banana", "mango", "orange")

print(fruits[-1])
print(fruits[-2])
```

```python
#tuple slicing 
fruits = ("apple", "banana", "mango", "orange", "grapes")

print(fruits[1:4])
```

```python
#slicing from begining 

print(fruits[:3])
```

```python
#slicing from end 

print(fruits[2:])
```

Tuples are immutable, which means their items cannot be changed after creation.

```python
#changing tuple indirectly 

fruits = ("apple", "banana", "mango")

fruit_list = list(fruits)
fruit_list[1] = "orange"

fruits = tuple(fruit_list)

print(fruits)
```

Dictionaries - dictionary is a collection that stores information in key–value pairs.

In Python, dictionaries are written using curly brackets {}.



```python
student = {
    "name": "Sadikshya",
    "course": "Python",
    "score": 90
}

print(student)
```

```python
#using dict 

student = dict(name="Sadikshya", course="Python", score=90)

print(student)
```

```python
#accessing dictionary values- Dictionaries do not normally use numerical indexes. Values are accessed using their keys.

student = {
    "name": "Sadikshya",
    "course": "Python",
    "score": 90
}

print(student["name"])
print(student["score"])
```

```python
#dictionaries are mutable : so their value can be changed

student = {
    "name": "Sadikshya",
    "score": 90
}

student["score"] = 95

print(student)
```
