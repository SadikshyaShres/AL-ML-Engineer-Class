```python
#Variables - used to store data 
name = "Sadikshya"
age= 29
height = 5.0
```

```python
#type of casting -Convert one data type into another.
age_str = str(age) #Interger to string 
height_int = int(height) #float to Interger
number = float(20) #Integer to float

print(age_str)
print(height_int)
print(number)
```

```python
#airthmetic operator- Perfomed mathmatical condition 
a = 10
b = 3

print(a + b) # Addition 
print(a - b) # Subtraction 
print(a * b) # Multiplication 
print(a / b) # Division 
print(a // b) # Floor Division 
print(a % b) # Modulus (Remainder) 
print(a ** b) # Exponent (Power)
```

```python
#comparison operator 
print(a == b) # Equal to 
print(a != b) # Not Equal to 
print(a > b) # Greater than 
print(a < b) # Less than 
print(a >= b) # Greater than or Equal to
print(a <= b) # Less than or Equal to
```

```python
#Assignment Operator-  Assign or update variable values.

x = 5 
x +=2  #x = x+2
x-=1 #x = x-1
x *=3 #x = x*3
x /=2  #x =x/2
 
print (x)
```

```python
#logical operator -Combine multiple conditions.
age = 23

print (age > 18  and age < 30)
print (age >30 and age ==23)
print(not(age>18))


```

```python
#Boolean- Stores only True or False.
is_student = True
is_graduated = False

print(is_student)
print(is_graduated)
```

```python
# Input Function- Takes input from the user.

name = input ("Enter your name:")
age = input ("Enter your age")
print ("My name is ", name)
print ("My age is ", age)
```

```python
#Branching (if-else)- Executes code based on a condition.

age = int(input("Enter your age: "))
if age >= 18: 
    print("You are eligible to vote.") 
else: 
    print("You are not eligible to vote.")
```

```python
#if-elif-else -Checks multiple conditions.

marks = int(input("Enter your marks: ")) 
if marks >= 90: 
    print("Grade A") 
elif marks >= 75: 
    print("Grade B")
elif marks >= 60: 
    print("Grade C") 
else: 
    print("Grade D")
```

```python
#Type Function # Returns the data type of a variable.
print(type(name))
print(type(age)) 
print(type(height))
```

```python
#Dynamic Typing- Same variable can store different data types.
value = 100 
value = "Python" 
value = 5.5 
print(value)
```

```python
# Strong Typing -Python does not automatically mix incompatible data types.Type casting is required.
num = 10 text = "20" # print(num + text) # Error 
print(num + int(text)) # Correct after type casting
```
