What Is an Iterator?

An iterator is an object that returns one value at a time.

The iter() function converts an iterable into an iterator.

The next() function gets the next value.

```python
numbers = [10, 20, 30]

number_iterator = iter(numbers)

print(next(number_iterator)) #The next() function asks an iterator for its next available item.
print(next(number_iterator))
print(next(number_iterator)) #Calling next() again produces a StopIteration error because there are no more values.
```

3. What is generator ?

A generator is a special type of function that produces values one at a time instead of returning all the values at once.

A generator uses the yield keyword.



```python
def generate_numbers():
    yield 1
    yield 2
    yield 3

for number in generate_numbers():
    print(number)
```

```python
def count_up_to(limit):
    number = 1

    while number <= limit:
        yield number
        number += 1

for value in count_up_to(5):
    print(value)
```

. Why Use yield Instead of return?

return ends the function and sends a value back.

yield pauses the function, sends a value, and remembers the function’s current state. The function can resume later from the same position.

```python
def generate_cubes(n):
    for number in range(n):
        yield number ** 3

        
```

```python
cubes = generate_cubes(5)

print(next(cubes))
print(next(cubes))
print(next(cubes))
```

```python
#Fibonacci Sequence Generator - The Fibonacci sequence begins with 0 and 1. Each following number is the sum of the two previous numbers.

def generate_fibonacci(n):
    first = 0
    second = 1

    for count in range(n):
        yield first

        next_number = first + second
        first = second
        second = next_number
```

```python
fibonacci = generate_fibonacci(7)

for number in fibonacci:
    print(number)
```

```python
#generator and iterator together - A generator is automatically an iterator. Therefore, we can directly use next() with a generator object.

def count_up_to(limit):
    number = 1

    while number <= limit:
        yield number
        number += 1

counter = count_up_to(3)

print(iter(counter) is counter)
print(next(counter))
print(next(counter))
print(next(counter))

```

4. MAP FUNCTION 

The map() function applies a function to every item in an iterable.

It requires:

A function
An iterable or sequence

```python
def square(number):
    return number ** 2

numbers = [1, 2, 3, 4]

squared_numbers = map(square, numbers)

print(list(squared_numbers))

#map() returns a map object, which is an iterator. We commonly convert it into a list to view all its values.
```

Lambda Function:

A lambda function is a small anonymous function. Anonymous means that the function does not require a regular name.

```python
#Regular function 

def square(number):
    return number ** 2
```

```python
square = lambda number: number ** 2

print(square(5))
```

```python
numbers = [1, 2, 3, 4]

squares = map(lambda number: number ** 2, numbers)

print(list(squares))
```

Filter Function 

The filter() function selects items from an iterable that meet a condition.

If the function returns True, the item is included. If it returns False, the item is excluded.

```python
def is_even(number):
    return number % 2 == 0

numbers = [1, 2, 3, 4, 5, 6]

even_numbers = filter(is_even, numbers)

print(list(even_numbers))
```

Reduce () Function - The reduce() function combines all the values in an iterable to produce one final value.

Unlike map() and filter(), reduce() must be imported from the functools module.

```python
from functools import reduce

numbers = [1, 2, 3, 4]

total = reduce(
    lambda first, second: first + second,
    numbers
)

print(total)
```
